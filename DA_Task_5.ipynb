```python
# --------------------------------------------
# 2. Load Dataset
# --------------------------------------------
df = pd.read_csv('train.csv')    # Load Titanic dataset (replace with your file if needed)
```

```python
# Step 3.1: View the first 5 rows of the dataset
print(df.head())

# Step 3.2: Show the data types and non-null counts for each column
print(df.info())

# Step 3.3: Get basic statistics for numerical columns
print(df.describe())

# Step 3.4: Check how many missing values are in each column
print(df.isnull().sum())

# Step 3.5: Check how many male vs female passengers are in the dataset
print(df['Sex'].value_counts())
```

```python
# Step 4.1: Plot the distribution of passenger age

df['Age'].hist(bins=20, edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Step 4.2: Plot a boxplot to detect outliers in fare

sns.boxplot(x=df['Fare'])
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.show()
```

```python
# Step 5.1: See how Age relates to Fare paid

sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Step 5.2: Compare average age of survived (1) vs not survived (0)

sns.barplot(x='Survived', y='Age', data=df)
plt.title('Survival by Average Age')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Average Age')
plt.show()
```

```python
# Step 6.1: Fix - Select only numeric columns before plotting heatmap

numeric_df = df.select_dtypes(include=['number'])  # Select only numeric columns
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

```
